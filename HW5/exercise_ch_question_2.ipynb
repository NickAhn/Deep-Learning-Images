```python
# Input and output vectors are given. 
inp = [ 0.7300, -1.0400, -1.2300,  1.6700, -0.6300,  1.4300, -0.8400,  0.1500,
         -2.3000,  3.1000, -1.4500, -1.8100,  1.8700, -0.1100, -0.2800,  1.1200,
         -0.4200,  2.8900]
out = [ 1.43,  10.1,  8.3,  1.03,  10.21, -0.1,  8.92,  5.1,
         -7.53, 34.72,  7.61,  3.2,  2.19,  7.15,  7.69, -0.18,
          8.81, 23.1]
```

```python
# Define the polynomial model of degree 3, i.e., having 3 weights and 1 bias. 
# Also define the loss function


# t = input tensor
# w, b = weights, bias
def model(t, w1, w2, w3, b):
    return (t**3 * w1) + (t**2 * w2) + (t * w3) + b


# pred = t_c  ->  predicted output
# act = t_p  ->  actual output
def loss_fn(pred, act):
    diff = (pred - act)**2
    return diff.mean()
```

```python
# Define gradient manually wrt the exisiting parameters
# Note: You need to define appropriate derivative functions to define the gradient
# Use the defined gradient function to define the training function 
# Note: You cannot use autograd and optimizers
# Run it on the input and output vector with appropriate learning rate and number of iterations
# Plot the learned curve

from scipy.interpolate import interp1d
import numpy as np
def dloss_fn(pred, act):
    dsq_diffs = 2*(pred - act) / pred.size(0)
    return dsq_diffs

# 3 * t**2 * w1 + 2 * t * w2 + 1 * t_u
def dmodel_dw1(t, w1, w2, w3, b):
    return 3 * w1 * t**2

def dmodel_dw2(t, w1, w2, w3, b):
    return 2 * t * w2

def dmodel_dw3(t, w1, w2, w3, b):
    return t

def dmodel_db(t, w1, w2, w3, b):
    return 1.0


# t_c = predicted
# t_p = actual
def grad_fn(t, t_c, t_p, w1, w2, w3, b):
    dloss_dtp = dloss_fn(t_p, t_c)
    dloss_dw1 = dloss_dtp * dmodel_dw1(t, w1, w2, w3, b)
    dloss_dw2 = dloss_dtp * dmodel_dw2(t, w1, w2, w3, b)
    dloss_dw3 = dloss_dtp * dmodel_dw3(t, w1, w2, w3, b)
    dloss_db = dloss_dtp * dmodel_db(t, w1, w2, w3, b)
    return torch.stack([
        dloss_dw1.sum(),
        dloss_dw2.sum(),
        dloss_dw3.sum(),
        dloss_db.sum()
    ])


# t_input = t_u  ->  input tensor
# t_pred = t_c  ->  predicted
# t_act = t_p  ->  actual
def training_loop(n_epochs, learning_rate, params, t_input, t_pred):
    for epoch in range(1, n_epochs+1):
        w1, w2, w3, b = params
        t_act = model(t_input, w1, w2, w3, b) #forward pass
        loss = loss_fn(t_pred, t_act)
        grad = grad_fn(t_input, t_pred, t_act, w1, w2, w3, b) #backwards pass
        
        params = params - learning_rate*grad
        print('Epoch %d, Loss %f' % (epoch, float(loss)))
        
    return params


import torch
%matplotlib inline
from matplotlib import pyplot as plt

params = torch.tensor([1.0, 1.0, 1.0, 0.0], requires_grad=True)
# Convert out to tensor
t_inp = torch.Tensor(inp) #t_u
t_out = torch.Tensor(out) #t_c

# Normalizing Tensors
# t_inp  = (t_inp - t_inp.mean()) / t_inp.std()
# t_out  = (t_out - t_out.mean()) / t_out.std() 

params = training_loop(n_epochs=10000, learning_rate=1e-3, params=params, t_input=t_inp, t_pred=t_out)
t_p = model(t_inp, *params) # actual


fig = plt.figure(dpi=600)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(t_inp.numpy(), t_p.detach().numpy(), 'o')
plt.plot(t_inp.numpy(), t_out.numpy(), 'o')

params
```

```python
# Use PyTorch's autograd to automatically compute the gradients 
# Define the training function
# Note: You cannot use optimizers.
# Run it on the input and output vector with appropriate learning rate and number of iterations
# Plot the learned curve
import torch

params = torch.tensor([1.0, 1.0, 1.0, 0], requires_grad=True)


# t_inp = t_u
# t_pred = t_c
# t_act = t_p
def training_loop_autograd(n_epochs, learning_rate, params, t_inp, t_pred):
    for epoch in range(1, n_epochs+1):
        if params.grad is not None:
            params.grad_zero()
        t_act = model(t_inp, *params)
        loss = loss_fn(t_act, t_pred)
        loss.backward()
        
        with torch.no_grad():
            params -= learning_rate * params.grad
            
        if epoch % 500 == 0:
            print('Epoch %d, Loss %f' % (epoch, float(loss)))
    return params

params = training_loop_autograd(
    100,
    0.01,
    params,
    t_inp,
    t_out
)
t_pred = model(t_in, *params)

fig = plt.figure(dpi=600)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(t_inp.numpy(), t_p.detach().numpy(), 'o')
plt.plot(t_inp.numpy(), t_out.numpy(), 'o')

params
```

```python
# Use PyTorch's autograd to automatically compute the gradients 
# Use optimizers to abstract how parameters get updated
# Define the training function
# Run it on the input and output vector with appropriate learning rate, number of iterations, and SGD optimizer
# Plot the learned curve
import torch.optim as optim

def training_loop_autograd(n_epochs, optimizer, params, t_u, t_c):
    for epoch in range(1, n_epochs+1):
        t_p = model(t_u, *params)
        loss = loss_fn(t_p, t_c)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
            
        if epoch % 500 == 0:
            print('Epoch %d, Loss %f' % (epoch, float(loss)))

learning_rate = 1e-5
params = torch.tensor([1.0, 1.0, 1.0, 0], requires_grad=True)
optimizer = optim.SGD([params], lr=learning_rate) # SGD = stochastic gradient descent
            
training_loop_autograd(
    n_epochs=500,
    optimizer=optimizer,
    params=params,
    t_u=t_u,
    t_c=t_c
)
t_p = model(t_inp, *params) # actual


fig = plt.figure(dpi=600)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(t_inp.numpy(), t_p.detach().numpy(), 'o')
plt.plot(t_inp.numpy(), t_out.numpy(), 'o')

params
```

```python
# Split the dataset to training set (80%) and validation set (20%) randomly
# Use PyTorch's autograd to automatically compute the gradients 
# Use optimizers to abstract how parameters get updated
# Define the training function that tracks both training and validation losses
# Run it on the input and output vector with appropriate learning rate, number of iterations, and SGD optimizer
# Plot the learned curve


def training_loop(n_epochs, optimizer, params, train_t_in, val_t_in, train_t_out, val_t_out):
    for epoch in range(1, n_epochs + 1):
        train_t_pred = model(train_t_in, *params)
        train_loss = loss_fn(train_t_pred, train_t_out)
        val_t_pred = model(val_t_in, *params)
        val_loss = loss_fn(val_t_pred, val_t_out)
        
        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()
        
        if epoch <= 3 or epoch % 500 == 0:
            print(f"Epoch {epoch}, Training loss {train_loss.item():.4f},"
                  f" Validation loss {val_loss.item():.4f}")
    return params

samples_num = t_in.shape[0]
n_val = int(0.2 * samples_num)

shuffled_indices = torch.randperm(samples_num)

train_indices = shuffled_indices[:-n_val]
val_indices = shuffled_indices[-n_val:]

train_indices, val_indices

train_t_in = t_inp[train_indices]
train_t_out = t_out[train_indices]
val_t_in = t_inp[val_indices]
val_t_out = t_out[val_indices]

train_t_in = 0.1 * train_t_in
val_t_in = 0.1 * val_t_in

params = torch.tensor([1.0, 1.0, 1.0, 0.0], requires_grad=True)
learning_rate = 1e-3
optimizer = optim.SGD([params], lr=learning_rate)

training_loop(
    n_epochs = 3000,
    optimizer = optimizer,
    params = params,
    train_t_in = train_t_in,
    val_t_in = val_t_in,
    train_t_out = train_t_out,
    val_t_out = val_t_out
)
t_p = model(t_inp, *params)

fig = plt.figure(dpi=600)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(t_inp.numpy(), t_p.detach().numpy())
plt.plot(t_inp.numpy(), t_out.numpy(), 'o')
```
